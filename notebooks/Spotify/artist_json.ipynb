```python
import pandas as pd
import numpy as n
import json
import re
```

```python
DATA_DIR = "../data/"
RAW_DATA_DIR = DATA_DIR + "raw.csv"

JSON_DATA_DIR = DATA_DIR + "json/"
TEST_JSON_DATA_DIR = JSON_DATA_DIR + "test_pl.json"

# temporary
RAW_DATA_DIR = 'spotify_dataset.csv'
```

```python
#standart read csv won't work, commas are in names!!! 
n = 4 #number of columns
df = pd.read_csv(RAW_DATA_DIR,
                 usecols=range(n),
                 lineterminator='\n',
                 header=None)

df = df[1:] #take the data less the header row
df.columns = ["user_id", "artistname", "trackname", "playlistname"] #set the header row as the df header
```

```python
# data cleaning
df = df[df['artistname'].apply(lambda x: len(re.sub(r'[a-zA-Z0-9&.!\s\',-üéµ\(\)$+"*\'#]', '', str(x)))) == 0]
```

```python
art_gb = df.groupby("artistname")
# test_art = art_gb.get_group((list(art_gb.groups)[0]))
test_art = art_gb.get_group("Tame Impala")
test_art_name = "Tame Impala"
```

```python
display(test_art_name)
test_art
```

```python
master_json_list = {}
```

```python
test_art_us = test_art.user_id.unique()
test_art_us_cnt = len(test_art_us)


test_art_tr = test_art.trackname
test_art_pl = test_art.playlistname.unique()


test_art_json = {
        #"name": test_pl_name,
        "num_users": test_art_us_cnt,
        "num_adds": len(test_art_tr), # idk if this is relevant
        "num_tracks": len(test_art_tr.unique()),
        "num_pl": len(test_art_pl),
        "tracks": list(test_art_tr.unique()),
}
master_json_list[test_art_name] = test_art_json
```
