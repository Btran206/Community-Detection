```python
import pandas as pd
import numpy as n
import json
```

```python
DATA_DIR = "../data/"
RAW_DATA_DIR = DATA_DIR + "raw.csv"

JSON_DATA_DIR = DATA_DIR + "json/"
TEST_JSON_DATA_DIR = JSON_DATA_DIR + "test_pl.json"

RAW_DATA_DIR
```

```python
#standart read csv won't work, commas are in names!!! 
n = 4 #number of columns
df = pd.read_csv(RAW_DATA_DIR,
                 usecols=range(n),
                 lineterminator='\n',
                 header=None)

df = df[1:] #take the data less the header row
df.columns = ["user_id", "artistname", "trackname", "playlistname"] #set the header row as the df header
```

```python
df
```

```python
pl_gb = df.groupby("playlistname")
test_pl = pl_gb.get_group((list(pl_gb.groups)[1]))
test_pl_name = list(pl_gb.groups)[1]
```

```python
display(test_pl_name)
test_pl
```

```python
master_json_list = {}
```

```python
test_pl_us = test_pl.user_id.unique()
test_pl_us_cnt = len(test_pl_us)

test_pl_tr = test_pl.trackname.unique()
test_pl_ar = test_pl.artistname.unique()


test_pl_json = {
        #"name": test_pl_name,
        "collaborative": test_pl_us_cnt > 1,
        "num_users": test_pl_us_cnt, 
        "num_edits": len(test_pl), # here for if a song was added twice, check in EDA to see if this actually happens

        "num_tracks": len(test_pl_tr),
        "num_artists": len(test_pl_ar),
        "tracks": list(test_pl_tr),
        "artists": list(test_pl_ar)
}
master_json_list[test_pl_name] = test_pl_json
```

```python
master_json_list
```

```python
with open(TEST_JSON_DATA_DIR, "w") as outfile:
    json.dump(master_json_list, outfile)
```
